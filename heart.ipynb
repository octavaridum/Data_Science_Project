```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
#Removing duplicated values
df = df.loc[~df.duplicated()].copy()
```

* More EDA is required to understand the relationship between the features

## Data is ready for modeling

```python
df.columns
```

```python
# breaking deoendent and independet varaiable
x = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']]
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.30, random_state=101)
```

## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
#Cheking prediction of the model 
prediction = logmodel.predict(x_test)
prediction
```

```python
prediction_df = pd.DataFrame(prediction,columns=["target_predicted"])
prediction_df
```

```python
#Testing model accuracy
from sklearn.metrics import confusion_matrix
```

```python
accuracy = confusion_matrix(y_test,prediction)
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,prediction)
accuracy
```

## Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dtmodel = DecisionTreeClassifier()
```

```python
dtmodel.fit(x_train,y_train)
```

```python
#predicttion with decison tree model
prediction = dtmodel.predict(x_test)
prediction
```

```python
#Testing the accuracy of the model 
from sklearn.metrics import confusion_matrix
```

```python
accuracy = confusion_matrix(prediction,y_test)
accuracy
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(prediction,y_test)
accuracy
```

```python
rfclass.fit(x_train,y_train)
```

```python
prediction = rfclass.predict(x_test)
prediction
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(prediction,y_test)
accuracy
```
